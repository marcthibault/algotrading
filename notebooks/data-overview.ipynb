```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import datetime

data_all = pd.read_csv("../data/WIKI-data-all.csv").set_index(["date", "ticker"]).sort_index()
```

```python
data_all.head()
```

```python
open_tickers_per_date = data_all.groupby("date").open.count()
```

```python
plt.plot(open_tickers_per_date.loc["2005-01-01":])
plt.ylabel("number of open stocks")
plt.ylim((0, plt.ylim()[1]))
```

```python
# Python = Monday
def weekday(datestring):
    year, month, day = (int(x) for x in datestring.split('-'))
    return datetime.date(year, month, day).weekday()
    
```

```python
open_tickers_per_date_df = pd.DataFrame(open_tickers_per_date)
```

```python
open_tickers_per_date_df.loc[:, "weekday"] = [weekday(datestring) for datestring in open_tickers_per_date_df.index]
```

```python
for d in range(7):
    print(open_tickers_per_date_df.query("weekday == " + str(d)).open.mean())
```
